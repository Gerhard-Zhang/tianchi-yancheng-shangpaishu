```python
import numpy as np
import pandas as pd

# 获取全部数据
train = np.loadtxt('data/train_20171215.txt',dtype='int')
train = train.tolist()
train
```

```python
# 根据牌子分类，并分开保存
sale_with_car_type = [[],[],[],[],[]]

for i in range(0, len(train)):
    car_type = train[i][2] - 1
    sale_with_car_type[car_type].append(train[i])

for i in range(0,5):
    print(len(sale_with_car_type[i]))
```

```python
# 去掉类别一列
for i in range(0,5):
    sale_with_car_type[i] = np.delete(sale_with_car_type[i],2,axis=1)
    print(sale_with_car_type[i])
```

```python
# 画图
import matplotlib.pyplot as plt
%matplotlib inline
for i in range(0,5):
    pro_train = sale_with_car_type[i]
    x = pro_train[:,0]
    y = pro_train[:,2]

    x = x[0:150]
    y = y[0:150]

    plt.figure(dpi=100)
    plt.plot(x,y)
```
